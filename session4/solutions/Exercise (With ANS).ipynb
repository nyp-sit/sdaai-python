## Task 21. Merge the two dataframes, taking the second dataframe(df2) as additional updates

````python
merged_df = pd.merge(df, df2, on='tender_no', how='outer')

for col in df.columns:
    if col != 'tender_no':  # Exclude the 'tender_no' column itself
        merged_df[col] = merged_df[col + '_x'].fillna(merged_df[col + '_y'])

merged_df.drop(columns=[col for col in merged_df.columns if '_x' in col or '_y' in col], inplace=True)
merged_df
````

## Task 25. Create a column that flags out "awarded_amt" more than or equal to 1 billion for further review (using ```apply```)

````python
def further_review(amount):
    if amount > 1000000000:
        return 'High-value'
    else:
        return 'Normal'
````
