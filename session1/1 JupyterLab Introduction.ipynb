## Understanding cell types

Within the notebook are cells defined as Code, Markdown or Raw.

The **Code** cells are where the we will input the Python instructions. When the cell is run, the result is display in an output cell.

The **Markdown** cells will provide the documentation to the codes within the notebook. Text in these cells will be formatted using markdown language, which allows for formatting features such as **bold** and *italic*.

The **Raw** cells are not evaluated by the notebook kernel. These cells are used to render the notebook into another format such as HTML or LaTeX through nbconvert.

### **Bold** and *italics* 

You can emphasize text using **bold** or *italics*. To do so, wrap the text between asterisks ```*``` or underscores ```_```.

If you are intending to use the asterisk or underscore without emphasising, you can always escape with a backslash ```\```, eg ```\*``` or ```\_```.


#### Markdown
```
This is in _italics_. \
This is in *italics*. \
This is in __bold__. \
This is in **bold**. \
This is in ___bold italics___. \
This is in ***bold italics***.

Using Markup tags works for <i>italics</i>, <b>bold</b>, and <b><i>bold italics</i></b> too.
```

#### Formatted display
This is in _italics_. \
This is in *italics*. \
This is in __bold__. \
This is in **bold**. \
This is in ___bold italics___. \
This is in ***bold italics***.

Using Markup tags works for <i>italics</i>, <b>bold</b>, and <b><i>bold italics</i></b> too.

### Keyboard Shortcuts
The need to switch between using the mouse and the keyboard generally increases the time spent to complete a task. Therefore, it is always welcomed to have keyboard shortcuts for frequent tasks that you have to do in a notebook.

Familiarizing yourself with some of these shortcuts can greatly improve your efficiency.

The shortcuts can also be modified based on your preference if necessary



##### Shortcuts when in command mode

* ```i```, ```i``` (Interrupt the kernel in case something is taking too long) 
* ```0```, ```0``` (\[*zero,zero*\] Restart the kernel that is needed to run the code)
* ```a``` (Insert a new cell above cell selection)
* ```b``` (Insert a new cell below cell selection)
* ```d```, ```d``` (Delete selected cells)

* ```x``` (Cut selected cells)
* ```c``` (Copy selected cells)
* ```v``` (Paste cells below selection)
* ```z``` (Undo cell operation)
* ```y``` (Change cell type to Code)
* ```m``` (Change cell type to Markdown)

* ```Shift``` + ```Up``` (Select current selection and the cell above it)
* ```Shift``` + ```Down``` (Select current selection and the cell below it)
* ```Shift``` + ```m``` (Merges the content of the selected cells into 1 cell)

##### Shortcuts when in edit mode

* ```Ctrl``` +```a``` (Select all)
* ```Ctrl``` +```z``` (Undo)
* ```Ctrl``` +```Left``` (Go one word left)
* ```Ctrl``` +```Right``` (Go one word right)
* ```Ctrl``` +```Shift``` + ```-``` (Split the content in the current cell into 2 cells based on the location of the caret)
